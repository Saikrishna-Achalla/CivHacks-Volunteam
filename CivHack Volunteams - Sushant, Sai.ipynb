```python
import pandas as pd
import numpy as np
import random
```

```python
genz_names = ["Tom", "James", "Khoi", "John", "Jack", "Sean", "Tim", "Jake"]

hobbies = ["Running", "Reading", "Tennis", "Yoga", "Travelling", "Teaching", "Board Games", "Meditation", 
                "Painting", "Guitar", "Cello", "Violin", "Piano", "Gardening", "Music", "Television", "Fishing",
               "Cooking"]

locations = ["Berkeley", "Austin", "New York", "Princeton", "San Diego", "Santa Clara", "Los Angeles", 
                  "Houston"]





def one_row(names, hobbies, locations):
    
    hobbies_arr = random.sample(hobbies, 3)
    name = random.sample(names, 1)
    location = random.sample(locations, 1)
    array = np.concatenate((name, location, hobbies_arr), axis = None)
    return array


def df_creation_genz(names, hobbies, locations):
    df = pd.DataFrame(columns=['Name', 'Location', "Hobby 1", "Hobby 2", "Hobby 3", "Age", "Hours"])
    
    for i in range(100):
        value = random.randint(18, 31)
        hours = random.randint(1, 10)
        one = one_row(names, hobbies, locations)
        values_to_add = {'Name': one[0], 'Location': one[1], "Hobby 1": one[2], "Hobby 2": one[3], "Hobby 3": one[4], "Age": value, "Hours": hours}
        row_to_add = pd.Series(values_to_add, name=i)
        df = df.append(row_to_add)
    return df



genz = df_creation_genz(genz_names, hobbies, locations)
genz
```

```python
def getScore(user, df):
    filtered_df = df[df["Location"] == user[1]]
    scores = []
    for index, row in filtered_df.iterrows():
        score = 0
        hobbies = [row['Hobby 1'],row['Hobby 2'],row['Hobby 3']]
        if(user[2] in hobbies):
            score += 20
        if(user[3] in hobbies):
            score += 20
        if(user[4] in hobbies):
            score += 20
        score += 45 - abs(user[-1] - row["Hours"])
        scores.append(score)
    filtered_df["Scores"] = scores
    filtered_df.sort_values("Scores",ascending = False)
    return filtered_df[filtered_df.columns[:-1]].head(1)
```
